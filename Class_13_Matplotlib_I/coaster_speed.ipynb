```python
%matplotlib notebook
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
x_axis = np.arange(0, 130, 10)

danger_drop = [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]

rail_gun = [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]
```

```python
plt.title("Coaster Speed Over Time")
plt.xlabel("Coaster Runtime")
plt.ylabel("Speed (MPH)")

plt.grid()
plt.xlim(min(x_axis), max(x_axis))
plt.plot(x_axis, danger_drop, label='Danger Drop', color="red")
plt.plot(x_axis, rail_gun, label='RailGun', color="green")
plt.legend()

plt.savefig("CoasterSpeed.png")
```
