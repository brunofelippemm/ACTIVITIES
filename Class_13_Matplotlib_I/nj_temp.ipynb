```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x_axis = np.arange(1,13,1)
fahre_avg_monthly_temp = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
celsius_avg_monthly_temp = [(f-32)*0.56 for f in fahre_avg_monthly_temp]


# for f in fahre_avg_monthly_temp:
#     celsius_avg_monthly_temp.append(round((f-32)*0.56, 2))

plt.plot(x_axis, fahre_avg_monthly_temp, label="Fahrenheit")
plt.plot(x_axis, celsius_avg_monthly_temp, label="Celsius")

plt.legend()
plt.xlabel("Months")
plt.ylabel("Temperature")
plt.show()
```
