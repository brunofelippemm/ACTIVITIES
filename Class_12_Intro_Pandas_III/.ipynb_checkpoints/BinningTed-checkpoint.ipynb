```python
# Import Dependencies
import pandas as pd
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
# Create a path to the csv and read it into a Pandas DataFrame
csv_path = "ted_talks.csv"
ted_df = pd.read_csv(csv_path)

ted_df.head()
```

```python
# Figure out the minimum and maximum views for a TED Talk
max_views = ted_df["views"].max()
min_views = ted_df["views"].min()

bin_range = (max_views - min_views)/10
bin_range
```

```python
# Create bins in which to place values based upon TED Talk views
bins = [min_views, min_views+bin_range, min_views+2*bin_range, min_views+3*bin_range, min_views+4*bin_range,
        min_views+5*bin_range, min_views+6*bin_range, min_views+7*bin_range, min_views+8*bin_range, max_views]

# Create labels for these bins
scores = ["2", "3", "4", "5", "6", "7", "8", "9", "10"]
```

```python
# Slice the data and place it into bins
ted_df["Score"] = pd.cut(ted_df["views"], bins, labels=scores)
ted_df.head()
```

```python
# Create a GroupBy object based upon "View Group"
grouped_ted_df = ted_df.groupby("Score")

# Find how many rows fall into each bin
scores_summary = pd.DataFrame(data=grouped_ted_df["title"].count().tolist(), columns=["Number of videos"], index=[scores])
scores_summary

# Get the average of each column within the GroupBy object
avg_df = grouped_ted_df.mean()
avg_df
```
