```python
# Dependencies
from matplotlib import pyplot as plt
import statistics as stats
import numpy as np
import pandas as pd
from scipy.stats import sem
```

```python
# Read and shuffle data
housing_data = pd.read_csv("housing_data.csv", header=None)
housing_data = housing_data.sample(frac=1).reset_index(drop=True)
housing_data.rename(columns={0:"CRIM", 1:"ZN", 2:"INDUS", 3: "CHAS",4: "NOX",5: "RM", 
                                            6:"AGE",7: "DIS",8: "RAD",9: "TAX",10: "PTRATIO",
                                            11:"B",12: "LSTAT",13: "MEDV"}, inplace=True)
housing_data.head()
```

```python
# Create a bunch of samples, each with div items
div = 20
lim = len(housing_data) // div
samples = [housing_data.iloc[(i * div):(i * div + div), 13]
           for i in range(0, lim)]
```

```python
means = [s.mean() for s in samples]
sem = [s.sem() for s in samples]
means
sem
```
