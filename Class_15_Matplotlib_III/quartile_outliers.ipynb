```python
# Dependencies
import matplotlib.pyplot as plt
from stats import median
import numpy as np
import math
```

```python
### Data Points
arr = np.array([2.3, 10.2,11.2, 12.3, 14.5, 14.6, 15.0, 15.1, 19.0, 24.0])
arr
```

```python
# Find the median
arr_median = np.median(arr)
arr_median
```

```python
# Use numpy to create quartiles
q2_arr = np.quantile(arr, .50) 
q1_arr = np.quantile(arr, .25) 
q3_arr = np.quantile(arr, .75) 
```

```python
# Print the quartiles
print("Q2 quantile of arr : ", q2_arr) 
print("Q1 quantile of arr : ", q1_arr) 
print("Q3 quantile of arr : ", q3_arr) 
```

```python
# Calculate the interquartile range
iqr_arr = q3_arr - q1_arr
iqr_arr
```

```python
# Find lower boundary
# Q1 - 1.5 * IQR
lower_bound = q1_arr - 1.5*iqr_arr
lower_bound
```

```python
# Find upper boundary
# Q3 + 1.5 * IQR
upper_bound = q3_arr + 1.5*iqr_arr
upper_bound
```

```python
# Check for any lower outliers
lower_outliers = []
for i in arr:
    if i < lower_bound:
        lower_outliers.append(i)
lower_outliers
```

```python
# Check for any upper outliers
upper_outliers = []
for i in arr:
    if i > upper_bound:
        upper_outliers.append(i)
upper_outliers
```

```python
# Create box plot
plt.boxplot(arr, autorange=True)
plt.show()
```
