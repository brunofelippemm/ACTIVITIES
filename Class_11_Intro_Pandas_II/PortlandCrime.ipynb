```python
# Import Dependencies
import pandas as pd
```

```python
# Reference the file where the CSV is located

# Import the data into a Pandas DataFrame
df = pd.read_csv("crime_incident_data2017.csv")
df.head()
```

```python
# look for missing values
df.isnull().sum()
```

```python
# drop null rows
df_cleaned = df.dropna()
df_cleaned
```

```python
# verify counts
df_cleaned.isnull().sum()
```

```python
# Check to see if there are any values with mispelled or similar values in "Offense Type"
df_cleaned["Offense Type"].value_counts()
```

```python
# Combine similar offenses
df_cleaned["Offense Type"]=df_cleaned["Offense Type"].replace({"Commercial Sex Acts": "Pornography/Obscene Material"})
```

```python
# Check to see if you comnbined similar offenses correctly in "Offense Type".
df_cleaned["Offense Type"].value_counts()
```

```python
# Create a new DataFrame that looks into a specific neighborhood
df_cleaned_downtown = df_cleaned.loc[df_cleaned["Neighborhood"] == "Downtown"]
```
