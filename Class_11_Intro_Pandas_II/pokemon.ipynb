```python
# Dependencies
import pandas as pd
```

```python
# File to load
```

```python
# Read with Pandas
df = pd.read_csv("Pokemon.csv")
df.head()
```

```python
# Extract the following columns: "Type 1", "HP", "Attack", "Sp. Atk", "Sp. Def", and "Speed"
clean_df = df[["Type 1", "HP", "Attack", "Sp. Atk", "Sp. Def","Speed"]]
clean_df.head()
```

```python
# Create a dataframe of the average stats for each type of pokemon.
type_avg_df = clean_df.groupby(["Type 1"]).mean()
type_avg_df.head()
```

```python
# Calculate the total power level of each type of pokemon by adding all of the stats together.
# Place the results into a new column.
type_avg_df["Total Power"] = type_avg_df["HP"]+type_avg_df["Attack"]+type_avg_df["Sp. Atk"]+type_avg_df["Sp. Def"]+type_avg_df["Speed"]
type_avg_df
```

```python
# Bonus: Sort the table by strongest type and export the resulting table to a new CSV.
strongest_pokemon = type_avg_df.sort_values(["Total Power"], ascending=False)
strongest_pokemon.reset_index(inplace=True)
strongest_pokemon
```
