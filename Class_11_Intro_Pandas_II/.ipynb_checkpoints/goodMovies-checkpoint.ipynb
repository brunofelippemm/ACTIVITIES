```python
# Dependencies
import pandas as pd
```

```python
# Load in File from resources
# 'movie_scores.csv'
```

```python
# Read and display with pandas
df = pd.read_csv("movie_scores.csv")
df.head()
```

```python
# List all the columns the table provides
for col in df.columns:
    print(col)
```

```python
# only care about Imdb, so create a new table that takes the Film and all the columns relating to IMDB

imdb_df = df[["FILM", "IMDB", "IMDB_norm", "IMDB_norm_round", "IMDB_user_vote_count"]]
imdb_df.head()
```

```python
# List only movies whose ratings are over 7 (out of 10) in IMDB
imdb_df_high_rating = imdb_df.loc[imdb_df["IMDB"] > 7]
imdb_df_high_rating.head()
```

```python
# Find lesser-known movies to watch, with fewer than 20K votes
lesser_know_high_rating = imdb_df_high_rating.loc[imdb_df_high_rating["IMDB_user_vote_count"] < 20000]
lesser_know_high_rating.head()
```

```python
# Finally, export this file to an Excel spreadsheet -- without the DataFrame index.
imdb_df.to_excel("IMDB_movies_dataset.xlsx", index=False)
```
